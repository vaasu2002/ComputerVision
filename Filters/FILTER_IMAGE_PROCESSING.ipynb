```python
img = cv2.imread('/content/image-58.png')

gaussian_kernel = np.array([[1/16, 1/8, 1/16],   #3x3 kernel
                [1/8, 1/4, 1/8],
                [1/16, 1/8, 1/16]])

conv_using_cv2 = cv2.filter2D(img, -1, gaussian_kernel, borderType=cv2.BORDER_CONSTANT) 
gaussian_using_cv2 = cv2.GaussianBlur(img, (3,3), 0, borderType=cv2.BORDER_CONSTANT)

conv_using_cv2 == gaussian_using_cv2
```

```python
from skimage import io, img_as_float
from skimage.filters import gaussian

img = img_as_float(io.imread('/content/image-58.png', as_gray=True))

gaussian_using_cv2 = cv2.GaussianBlur(img, (3,3), 0, borderType=cv2.BORDER_CONSTANT)

gaussian_using_skimage = gaussian(img, sigma=1, mode='constant', cval=0.0) # padding by adding a layer with constant term(mode='constant') and value of that constant term will be (cval=0.0)
```

###  **Median Denoising**

**->** Help full for removal of salt and pepper noises.

```python
import cv2
import numpy as np

image = cv2.imread('/content/image-58.png',0)
median_using_cv2 = cv2.medianBlur(image, 3)

plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.imshow(median_using_cv2)
```
